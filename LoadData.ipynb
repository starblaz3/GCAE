```python
# File descriptors
xml_file = "SemEval2014Task4/Restaurants_Train_v2.xml"
output_file = "data/se_2014_acsa_train.csv"
```

```python
''' Function whose input is the SemEval XML file and filters the data based on aspect term or 
    category and saves it in format useful for the function defined further '''
def filter_data(xml_file, output_file, classification_type="c", term="food"):
    indexer1 = ""
    indexer2 = ""
    indexer3 = ""
    if classification_type == "c": # If aspect category
        indexer1 = "aspectCategories"
        indexer2 = "aspectCategory"
        indexer3 = "@category"
    elif classification_type == "t": # If aspect term
        indexer1 = "aspectTerms"
        indexer3 = "@term"
        indexer2 = "aspectCategory"
    string=""
    with open(xml_file,'r') as f:
        string = f.read()
    o = xd.parse(string)
    j=json.dumps(o)
    json1_data = json.loads(j)
    type(json1_data)
    data = pd.DataFrame(json1_data['sentences']['sentence'])
    aspect_df = data.loc[:,["text",indexer1]]
    aspect_df = aspect_df.dropna()
    final_filter = []
    for i in range(len(aspect_df)):
        x = aspect_df.iloc[i,1][indexer2]
        if isinstance(x, list):
            for j in x:
                if j[indexer3] == term:
                    final_filter.append([aspect_df.iloc[i,0],j["@polarity"]])
        else:
            final_filter.append([aspect_df.iloc[i,0],x["@polarity"]])
    final_df = pd.DataFrame(final_filter, columns=["review","sentiment"])
    final_df.to_csv(output_file, index=None)
    return final_df
```

```python
filter_data(xml_file, output_file)
```
